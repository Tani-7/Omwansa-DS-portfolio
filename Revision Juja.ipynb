```python
import pandas as pd
import numpy as np
ages = [20,22,25,27,29,21,23,31,62,67,45,32]
bins = [18,25,35,60,100]
cats = pd.cut(ages,bins)
cats
```

```python
pd.value_counts(cats)
```

```python
pd.cut(ages,[18,26,35,61,100])
```

```python
group_names  = ['youth','youngadult','middleaged','senior']
```

```python
pd.cut(ages,[18,26,35,61,100], labels=group_names)
```
